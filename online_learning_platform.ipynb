```python
# Age Distribution

plt.figure(figsize=(5,3))
sns.histplot(df['age_group'], bins=30, color='#c8fc44')
plt.title('Age Distribution')
plt.show()
```

```python
# Device Usage
plt.figure(figsize=(5,3))
sns.countplot(x='learning_device',data=df ,color='#8dcbeb')
plt.title('Device Usage')
plt.xlabel('Device')
plt.ylabel('count')
plt.show()
```

```python
# Platform Analysis
plt.figure(figsize=(5,3))
sns.countplot(df['primary_platform'], color='#f5d987')
plt.title('Platform Analysis')
plt.show()
```

```python
# Role Analysis
plt.figure(figsize=(5,3))
sns.countplot(df['role'], color='#e0a5f2')
plt.title('Role Analysis')
plt.show()
```

```python
# Learning Purpose
plt.figure(figsize=(5,3))
sns.countplot(df['learning_purpose'])
plt.title('Learning Purpose')
#plt.xticks(rotation=30)
plt.show()
```

```python
# Satisfaction Score
plt.figure(figsize=(5,3))

sns.histplot(df['satisfaction_score'],bins=10,color='#edad68')
plt.title('Satisfaction Score')
plt.show()
```

```python
# Role vs Hours Per Week
plt.figure(figsize=(6,4))
sns.barplot(x='role', y='hours_per_week',data=df, color='#80cfaa')
plt.title('Role vs Hours Per Week')
plt.xlabel('role')
plt.ylabel('hours_per_week')
plt.show()
```

```python
# Age vs Primary Platform
plt.figure(figsize=(5,3))
sns.barplot(y='age_group', x='primary_platform', data=df , color='#aba7e8')
plt.title('Age vs Primary Platform')
plt.xlabel('Age')
plt.ylabel('primary_platform')
plt.xticks(rotation =90)
plt.show()
```

```python
# Time Spent vs Completion Rate
plt.figure()
sns.barplot(x='hours_per_week', y='completion_rate_percent',data=df,color='#f2715a')
plt.title('Time Spent vs Completion Rate')
plt.show()
```

```python
# Platform vs Satisfaction
plt.figure(figsize=(5,3))
plt.figure()
sns.boxplot(x='primary_platform', y='satisfaction_score',data=df, color='#60a697')
plt.title('Platform vs Satisfaction')
plt.xticks(rotation=45)
plt.show()
```

```python
# Purpose vs Study Time
plt.figure(figsize=(6,4))
sns.boxplot(x='learning_purpose', y='hours_per_week', data=df)
plt.title('Purpose vs Study Time')
plt.xticks(rotation=45)
plt.show()
```

```python
# Completion Rate by Device
plt.figure(figsize=(5,3))
sns.barplot(x='completion_rate_percent', y='learning_device',data=df,color='#98ab61')
plt.title('Completion Rate by Device')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(5,3))
plt.figure()
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Outliers in Hours Per Week
plt.figure(figsize=(5,3))
plt.figure()
sns.boxplot(x=df['hours_per_week'], color='#a8d2ed')
plt.title('Outliers in Hours Per Week')
plt.show()
```

```python
# Platform Popularity
plt.figure(figsize=(6,4))
df.groupby('learning_purpose')['completion_rate_percent'].mean().plot(kind='bar')
plt.title('Avg Completion Rate by Purpose')
plt.show()
```

```python
# Platform Popularity
plt.figure(figsize=(5,3))
df['primary_platform'].value_counts().plot(kind='bar',color='#e8e680')
plt.title('Platform Popularity')
plt.show()
```
